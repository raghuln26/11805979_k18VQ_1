```python
import tkinter as Tk       #--->the first three lib to work with tkinter
from tkinter import *
from tkinter import ttk    #-----> module contain notebook widget
#lib to work with varity of nlp operation like pos..etc
from textblob import TextBlob      
#module for scrolling the text
from tkinter.scrolledtext import *
from tkinter import filedialog    #---> module with open dialog function
from textblob import Word

#creating the parent window(root)
root = Tk()
#adding the title to the parent window
root.title("sentiment analysis GUI")
#adding the dimension to the parent window
root.geometry('1000x1000')
#creating tab control
tabControl = ttk.Notebook(root)
#creating the tab
tab1 = ttk.Frame(tabControl)
tab2 = ttk.Frame(tabControl)
#to create scrollbar
w = Scrollbar(root)
w.pack(side = RIGHT,fill = Y)


#adding the tab
tabControl.add(tab1,text="text file")

tabControl.add(tab2,text="text file")

#to make tab contol visible
tabControl.pack(expand=1,fill="both")

#creating label
#label used to diplay the text on the screen.
#grid() method is used to split the parent window into rows and column
label1 = Label(tab1,text="Description",padx=4, pady=4)
label1.grid(column=0,row=0)
label1 = Label(tab2,text="file processor",padx=4, pady=4)
label1.grid(column=0,row=0)

description = Label(tab1,text="intelligent sentiment analysis system using GUI",pady=5,padx=5)
description.grid(row=8,column=8,)


#function
#tk.filediag is module with open and save dialog function
#askopnfilename is to open file dialog that reqursts selection of an existing file.
def files():
    #read the file
    f1 = tk.filedialog.askopenfilename(filetypes=(("Text Files",".txt"),("All files","*")))
    r_text = open(f1).read()
    processed_file.insert(tk.END,r_text)
    
#get() method is used to return the information on the specified range(1.0 means starting)
#.insert() method is used to insert the specified string the given place show.
#new_text.lower will convert the information inside the new_text into lower case
def lower():
    #convert the text into lower caser
    text = processed_file.get('1.0',tk.END)
    new_text = TextBlob(text)
    processed_text = new_text.lower()
    result = '\n\n lower case: {}'.format(processed_text)
    show.insert(tk.END,result)
    
#new_text.words function will tokens the given text into words.
#and the split() method is used to split the words by("")
def tokensization():
    #tokens the text 
    text = processed_file.get('1.0',tk.END)
    new_text = TextBlob(text)
    processed_text = list(str(new_text.words).split(" "))
    result = '\n\ntokenised words:{}'.format(processed_text)
    show.insert(tk.END,result)
    
#.tag() used to display the text into parts of speech    
def parts_of_speech():
    #tag the text
    text = processed_file.get('1.0',tk.END)
    new_text = TextBlob(text)
    processed_text = new_text.tags
    result = '\n\ntags: {}'.format(processed_text)
    show.insert(tk.END,result)
    
#.lemmatize('v') convert into the text into base word that is here into verb
def lemmatization():
    #tag the text
    text = processed_file.get('1.0',tk.END)
    new_text = Word(text)
    processed_text = new_text.lemmatize('v')
    result = '\n\ntags: {}'.format(processed_text)
    show.insert(tk.END,result)

    
#.sentiment() is used to analysis the sentiment of the given text.
#and it will return the subjectivity and the polarity
#but the we used processed_txt.polarity so it will return polarity only
#and the in the will check wheather the sentiment is pos,neg or neutral 
def sentiment_analysis(*polarity):
    x=0
    if x<0:
        print("neg")
    elif x>0:
            print("pos")
    else:
                print("nautral")
    #analysis the text sentiment
    text = processed_file.get('1.0',tk.END)
    new_text = TextBlob(text)
    processed_text = new_text.sentiment
    x=processed_text.polarity
    sentiment_analysis(x)
    show.insert(tk.END,x)
    

    
#.delete method is used to delete the string in the specified range    
def erase_result():
    #this function delete the result in the show entry
    show.delete('1.0',END)
    

    
        

    
# FILE READING  AND PROCESSING TAB
l1=Label(tab2,text="Open the file")
l1.grid(row=1,column=1)

processed_file = ScrolledText(tab2)
processed_file.grid(row=2,column=0,columnspan=8,padx=4,pady=4)


#creating button
button1=Button(tab2,text="Open File", width=12,command = files,bg='#c5cae9')
button1.grid(row=3,column=0,padx=4,pady=4)

button2=Button(tab2,text="tokenzation", width=12,command = tokensization,bg='#c5cae9')
button2.grid(row=3,column=1,padx=4,pady=4)

button3=Button(tab2,text="pos tag", width=12,command = parts_of_speech,bg='#c5cae9')
button3.grid(row=3,column=2,padx=4,pady=4)

button4=Button(tab2,text="sentiment", width=12,command = sentiment_analysis,bg='#c5cae9')
button4.grid(row=3,column=3,padx=10,pady=10)

button6=Button(tab2,text="lower", width=12,command = lower ,bg='#c5cae9')
button6.grid(row=4,column=1,padx=10,pady=10)

button7=Button(tab2,text="close", width=12,command = root.destroy ,bg='#c5cae9')
button7.grid(row=4,column=2,padx=10,pady=10)

button8=Button(tab2,text="clear result", width=12,command = erase_result ,bg='#c5cae9')
button8.grid(row=4,column=3,padx=10,pady=10)

button9=Button(tab2,text="lemmatize", width=12,command = lemmatization ,bg='#c5cae9')
button9.grid(row=5,column=0,padx=10,pady=10)




#show is the place to display the text
show = ScrolledText(tab2)
show.grid(row=7,column=0,columnspan=8,padx=4,pady=4)
show.config(state="normal")





#run the application in the infinteloop
mainloop()
```
